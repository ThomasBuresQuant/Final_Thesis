## Regression simple X = PER

```python
import statsmodels.api as sm

reg_vars = ["Schiller_PER", "Total_return_t+1_Winsorisé"]
df_reg = df[reg_vars].dropna()  # <-- This step is required

x = sm.add_constant(df_reg[["Schiller_PER"]])
y = df_reg["Total_return_t+1_Winsorisé"]

model = sm.OLS(y, x).fit()

print("Intercept (const):", model.params["const"])
print("Slope (Schiller_PER):", model.params["Schiller_PER"])
print(model.summary2(float_format="%.6f"))
```

## Vérification des variables

```python
columns_to_check = [
    "DIVIDEND_PAYOUT_PER_SHARE",
    "Dividend",                       #log
    "DIVIDENDS_PER_SHARE",              #log
    "PRICE_TO_BOOK_VAL",
    "P/CSH_FLOW_RATIO",
    "RETURN_ON_EQUITY_-_TOTAL_(%)",
    "RETURN_ON_ASSETS",
    "OPERATING_PROFIT_MARGIN",
    "NET_MARGIN",
    "TOTAL_DEBT_%_COMMON_EQUITY"
]

for col in columns_to_check:
    count = df[col].notna().sum()
    print(f"{col}: {count} non-empty cells")
```

```python
def winsorize_series(series, lower=0.01, upper=0.99):
    """
    Winsorize a pandas Series at the specified lower and upper percentiles,
    leaving missing (NaN or inf) values unchanged.
    """
    # Remove inf, but leave NaN as-is
    series_clean = series.replace([np.inf, -np.inf], np.nan)
    low = series_clean.quantile(lower)
    high = series_clean.quantile(upper)
    # Only clip non-missing values
    return series_clean.where(series_clean.isna(), series_clean.clip(lower=low, upper=high))

for col in columns_to_check:
    df[f"{col}_winsorized"] = winsorize_series(df[col])

import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning)
```

```python
# List of Y variables to test
y_vars = [
    "Total_return_t+1_Winsorisé",
    "Total_return_t+3_Winsorisé",
    "Total_return_t+5_Winsorisé",
    "Total_return_t+7_Winsorisé",
    "Total_return_t+10_Winsorisé"
]

# Explanatory variables
X_cols = ["PER_Winsorisé_>0", "PRICE_TO_BOOK_VAL_winsorized"]

for y_var in y_vars:
    # Prepare the dataset (drop missing for X and Y)
    df_reg = df[X_cols + [y_var]].dropna()
    X = sm.add_constant(df_reg[X_cols])
    y = df_reg[y_var]
    
    # Fit the model
    model = sm.OLS(y, X).fit()
    
    print(f"Y = {y_var}")
    print(f"R²: {model.rsquared:.4f}")
    print(model.summary())
    print("\n" + "="*50 + "\n")
```

## Homoscedasticity

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Homoscedasticity >> voulu en nuage de point sans pattern
residuals = model.resid
fitted = model.fittedvalues
plt.scatter(fitted, residuals)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Fitted values")
plt.ylabel("Residuals")
plt.title("Homoscedasticity check")
plt.show()

# Breusch-Pagan test >> voulu supérieur à 0.05
from statsmodels.stats.diagnostic import het_breuschpagan
bp_test = het_breuschpagan(residuals, model.model.exog)
print('Breusch-Pagan p-value:', bp_test[1])
```

```python
relevant_variables = [
    "DIVIDEND_PAYOUT_PER_SHARE",
    "DIVIDENDS_PER_SHARE",              #log
    "PRICE_TO_BOOK_VAL",
    "P/CSH_FLOW_RATIO",
    "RETURN_ON_EQUITY_-_TOTAL_(%)",
    "RETURN_ON_ASSETS",
    "OPERATING_PROFIT_MARGIN",
    "NET_MARGIN",
    "TOTAL_DEBT_%_COMMON_EQUITY_winsorized",
    "log_MARKET_CAPITALIZATION_winsorized",
    "YEAR",
    "COMMON_DIVIDENDS_(CASH)",
    "TOTAL_INTANGIBLE_OT_ASSETS-NET",
    "P/CSH_FLOW_RATIO_winsorized",
    "ENTERPRISE_VALUE",
    "COMMON_SHAREHOLDERS'_EQUITY",
    "GROSS_INCOME",
    "NET_INCOME_BEFORE_PREFERRED_DI",
    "MINORITY_INTEREST.1",
    "EMPLOYEES",
    "NET_SALES_OR_REVENUES",
    "CAPITAL_EXPENDITURES",
    "DIVIDENDS_PROVIDED/PAID-COMMON",
    "EBIT_&_DEPRECIATION",
    "ACCOUNTS_PAYABLE",
    "CURRENT_LIABILITIES-TOTAL",
    "AMORTIZATION_OF_INTANGIBLES",
    "FUNDS_FROM_OPERATIONS",
    "DEPRECIATION_AND_DEPLETION",
]

import pandas as pd
import statsmodels.api as sm

def stepwise_selection(df, response, candidate_vars, initial_vars=[], 
                       threshold_in=0.1, verbose=True):
    included = list(initial_vars)
    while True:
        changed = False
        excluded = list(set(candidate_vars) - set(included))
        new_pvals = pd.Series(index=excluded, dtype=float)
        for new_var in excluded:
            model = sm.OLS(df[response], sm.add_constant(df[included + [new_var]])).fit()
            new_pvals[new_var] = model.pvalues[new_var]
        if new_pvals.empty:
            break
        best_pval = new_pvals.min()
        if best_pval < threshold_in:
            best_var = new_pvals.idxmin()
            included.append(best_var)
            changed = True
            if verbose:
                print(f"Add {best_var:30} with p-value {best_pval:.6f}")
        if not changed:
            break
    return included

# Variables to loop over
X1_options = ["PER_Winsorisé_>0", "Schiller_PER"]
Y_options = [
    "Total_return_t+1_Winsorisé",
    "Total_return_t+3_Winsorisé",
    "Total_return_t+5_Winsorisé",
    "Total_return_t+7_Winsorisé",
    "Total_return_t+10_Winsorisé"
]
candidate_vars = [col for col in relevant_variables if col not in X1_options]  # Exclude X1 from other variables

results = []

for Y in Y_options:
    for X1 in X1_options:
        # Prepare data (drop missing for all relevant columns)
        cols_needed = [Y, X1] + candidate_vars
        df_step = df[cols_needed].dropna()
        selected_vars = stepwise_selection(df_step, Y, candidate_vars, initial_vars=[X1], threshold_in=0.1, verbose=False)
        results.append({
            "Y": Y,
            "X1": X1,
            "selected_vars": selected_vars
        })
        print(f"\nY = {Y} | X1 = {X1}")
        print("Selected variables:", selected_vars)

# Results is a list of dictionaries with each configuration's selected variables
```
